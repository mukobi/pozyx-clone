Let's import the necessary libraries and modules first. "Params" at the end just let's helps us typeset math nicer for our plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt
from scipy.optimize import curve_fit
from math import pi
from physProps import rocket

params = {'mathtext.default': 'regular'}
plt.rcParams.update(params)
```

Now, let's define some physical laws that will be used throughout the script.

Note that "rocket" is a class defined in the physProps module (located in this directory), and everything preceding "rocket." is a physical property of the rocket defined in the class.

```python
g = -9.81 # gravity
Fg = g * rocket.mass # force of gravity
rho = 1.225 # density of air at sea level 20 degrees C
```

Here, we use pandas to import the data file in question. We select columns by their number, then rename them using "df.columns".

We use df.iloc to choose the datalines (we aren't interested in the data pre launch and post landing).

```python
df=pd.read_csv('rocket_launch.csv', delimiter=',', usecols = [0,1,28])

df.columns = ['index', 'time', 'range']


df2 = df.iloc[166:216].dropna() #index range pertaining to data of interest
```

For presentation purposes, we want to set the rocket's initial position and launch time to zero.

```python
xa=df2['time'].values
xi = xa[0] # when was the rocket launched?
x_final = xa[-1] - xi # sets land time assuming launch time is zero
x2 = xa - xi # sets launch time to zero

da=df2['range'].values*0.001 # convert from mm to m
di = da[0] # rocket's initial distance?
d_final = da[-1] - di # sets land distance assuming launch distance is zero
d2 = da - di # sets release time to zero
```

```python
df3 = df.ix[168:216].dropna() # for newRange calculations
x3 = df3['time'].values
raw_d3=df3['range'].values*0.001
d3 = raw_d3 - raw_d3[0]

dt = x3[-1] - x3[0]     # rocket's time in the air
s = d3[-1] - d3[0]      # final distance from anchor - initial distance from anchor
new_v = s / dt          # for discounting horizontal motion
new_x = new_v * dt      # for discounting horizontal motion

newd = np.sqrt(d2**2 - new_x**2) # total range discounting horizontal motion
plain_newRange = newd[~np.isnan(newd)] # drop any N
newRange = plain_newRange - plain_newRange[0] # set newRange to start at distance = 0


new_time = df2['time'].ix[167:216].values


adj_time = new_time - new_time[0]
```

```python
N = 100 # allows you to create each model out of 100 points
model_times = np.linspace(x2[0], x2[-1], 100)   # create array of 100 times within experimental timeframe for the modeled positions
x_mod = np.zeros(shape=(N))   # create empty array with space for 100 elements
v_mod = np.zeros(shape=(N))
a_mod = np.zeros(shape=(N))
dragForce = np.zeros(shape=(N))
dt = ( x2[-1]-x2[0] ) / N
```

```python
plt.figure()
plt.plot(x2,d2, 'C7', marker = 'o', linestyle="None")
```

```python
c_vect = np.array([0,0.01, 0.03, 0.05])
for c in range(4) :
    x_mod[0] = 0    # initial position of zero
    v_mod[0] = 17.79    # initial velocity of zero
    dragForce[0] = rocket.xSectionArea * 0.5 * c_vect[c] * rho * v_mod[0]**2   # initial drag force
    a_mod[0] = ( Fg-dragForce[0] ) / rocket.mass  # initial acceleration based on initial drag force
    n=0

    for n in range(0,N-1):
        x_mod[n+1] = x_mod[n]  + v_mod[n] * dt  + (1/2) * (a_mod[n]) * dt**2 # simply y=x + v*t + (1/2)*a*t^2
        v_mod[n+1] = v_mod[n] + a_mod[n] * dt # updated velocity is v + a*t

        if v_mod[n] < 0 :
            dragForce[n+1] = -( c_vect[c]*( (rho*v_mod[n]**2) /2 ) * rocket.xSectionArea ) # if/else may not be necessary

        else:
            dragForce[n+1] = c_vect[c]*( (rho*v_mod[n]**2) /2 ) * rocket.xSectionArea

        a_mod[n+1] = ( Fg-dragForce[n+1] ) /rocket.mass  # update acceleration with updated drag force

        n = n+1
            #plot the model results
    plt.plot(model_times, x_mod, label='$C_d={}$'.format(c_vect[c]))
    plt.plot(x2,d2, 'C7', marker = 'o', linestyle="None")
```

```python
plt.xlabel("Time (s)", fontsize=15)
plt.ylabel("Position (m)", fontsize=15)
plt.title("Position", fontsize=20)
plt.legend()
plt.tick_params(labelsize='large')

plt.show()
```
